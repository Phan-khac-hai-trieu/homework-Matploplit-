**5. Multiple line chart**

Create x vector which has 100 elements and the element values gradually increase from 0 to 99.

Randomly create y1 and y2 vector which have 100 elements and the element values are randomly generated in the range of \[0, 1000) 

Let draw the above two lines y1 and y2. The chart must has:


1.   names of x-axis and y-axis
2.   chart legend
3.   different colors for the lines of y1 and y2
4.   different markers for the lines of y1 and y2





```
x = np.arange(0,100)
y1 = np.random.randint( low = 0 , high = 1000 ,size = 100 )
y2 = np.random.randint( low = 500, high = 1000 , size = 100)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(y1,x,'bo:')
l2 = ax.plot(y2,x,'cD:')
ax.legend(labels = ('pc', 'console'), loc = 'lower center')
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```

**7. Draw a histogram**

```
smoker_man = np.array([ 190 , 31 ,70 ,140 , 50 , 60 ,55 , 150 , 181 , 209])
smoker_women = np.array([34, 100, 234, 43,55,123 , 125])
plt.hist([smoker_man,smoker_women], bins = [80,100,125,150] , rwidth=1, color=['red','blue'],label=['men','women'])
plt.title("proportions of smoker")
plt.xlabel('smoke range')
plt.ylabel('smoke habit')
plt.show()
```
